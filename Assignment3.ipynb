#### **Answer 4**

```python
def not_mutual_with_cap(hero_name):
  cap_friends = hero_network[hero_network['hero1'].str.contains("CAPTAIN AMERICA")]
  hero_friends = hero_network[hero_network['hero1'].str.contains(hero_name)]

  return hero_friends[np.logical_not(hero_friends['hero2'].isin(cap_friends['hero2']))]
```

```python
not_mutual_with_cap("SPIDER-MAN").describe()
```

We can see that spiderman has **738 unique** friends that are not friends of captain america.

#### **Answer 3A**

We sort the table to show earthquake over 5 in magnitude : 

```python
df_above_five_max = df[(df['Magnitude']) > 5 ]
df_above_five_max
```

```python
def find_year(date):
    if date != None:
      list_date = date.split("/")
      year = list_date[-1]
      return year

df_above_five_max['Year'] = df_above_five_max['Date'].apply(find_year)
df_above_five_max
```

We created a new column "Year" that contains only the year label from "Date" column DD/MM/YYYY with a function for every row in date. 

Next, we used groupby in order to show how many earthquake per year

```python
events_per_year = df_above_five_max.groupby("Year",as_index=False)["Date"].count()
events_per_year = events_per_year.rename(columns={'Date': 'Count'})
events_per_year
```

After it , we will find the max value in the table

```python
events_per_year[events_per_year['Count']==events_per_year['Count'].max()]
```

#### **Answer 3B**

We will sort the table for eartquake over 8 in magnitude

```python
earthquake_table = df[(df['Magnitude']) > 8 ]
earthquake_table
```

Like the previous question , we will add a column year from column Date

```python
def find_year(date):
    if date != None:
      list_date = date.split("/")
      year = list_date[-1]
      return year

earthquake_table['Year'] = earthquake_table['Date'].apply(find_year)
earthquake_table
```

We will use grouby to create a new table for count earthquake per year

```python
count_events = earthquake_table.groupby("Year",as_index=False)["Date"].count()
count_events = count_events.rename(columns={'Date': 'Count'})
count_events
```

```python
count_events[count_events['Count']==count_events['Count'].max()]
```

#### **Answer 4**

We will create a new column hour from "Time" column

```python
def find_hour(time):
    if time != None:
      list_time = time.split(":")
      hour = list_time[0]
      return hour

df['Hour'] = df['Time'].apply(find_hour)
df
```

We will use groupby in order to create a table for count events per hour, and find the max. 

```python
g = df.groupby("Hour",as_index=False)["Time"].count()
g = g.rename(columns={'Time': 'Count'})
g[g['Count']==g['Count'].max()]
```

# Part 3: FIFA 2019

```python
fifa = pd.read_csv('data.csv')
```

```python
fifa
```

#### **Answer 1**

```python
def strongest_player(nation):
  nation_players = fifa[fifa['Nationality'] == nation]
  max_overall = max(nation_players['Overall'])
  return nation_players[nation_players['Overall'] == max_overall]['Name']
```

```python
print("Strongest player in Brazil:")
print(strongest_player('Brazil'))
print("-------------------------------")
print("Strongest player in France:")
print(strongest_player('France'))
```

* The strongest player in Brazil is "Neyman Jr"
* The strongest players in France are "N. KantÃ©" and "A. Griezmann" (Both have the same and highest Overall rating)
